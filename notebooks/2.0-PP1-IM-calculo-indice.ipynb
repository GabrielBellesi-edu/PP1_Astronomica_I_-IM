```python
# # An√°lisis de √çndice de Masa de Meteoritos

# Este notebook utiliza la clase `MeteoriteMassAnalyzer` para calcular el √≠ndice de masa de meteoritos usando el m√©todo est√°ndar de regresi√≥n log-log.

# **Ubicaci√≥n:** `notebooks/2.0-PP1-IM-calculo-indice.ipynb`

# ## 1. Configuraci√≥n Inicial


# Importar librer√≠as necesarias
import sys
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

# Configurar matplotlib para mostrar gr√°ficos en la notebook
%matplotlib inline
plt.style.use('default')
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['font.size'] = 10

# Agregar el directorio ra√≠z del proyecto al path para importar el m√≥dulo
project_root = os.path.abspath('..')  # Subir un nivel desde notebooks/
if project_root not in sys.path:
    sys.path.append(project_root)

# Importar la clase principal
from calculo.calcular_indice_masa import MeteoriteMassAnalyzer

print("‚úì Librer√≠as importadas correctamente")
print(f"‚úì Directorio del proyecto: {project_root}")

# %% [markdown]
# ## 2. Inicializaci√≥n del Analizador

# %%
# Crear instancia del analizador
analyzer = MeteoriteMassAnalyzer(project_root=project_root)

# %% [markdown]
# ## 3. Exploraci√≥n Inicial de Datos

# %%
# Cargar datos para exploraci√≥n
print("Cargando datos para exploraci√≥n...")
amplitudes, df = analyzer.load_data(file_type='csv')

if amplitudes is not None:
    print(f"\nüìä ESTAD√çSTICAS B√ÅSICAS:")
    print(f"   ‚Ä¢ Total de registros: {len(amplitudes)}")
    print(f"   ‚Ä¢ Amplitud m√≠nima: {amplitudes.min():.2f}")
    print(f"   ‚Ä¢ Amplitud m√°xima: {amplitudes.max():.2f}")
    print(f"   ‚Ä¢ Amplitud promedio: {amplitudes.mean():.2f}")
    print(f"   ‚Ä¢ Desviaci√≥n est√°ndar: {amplitudes.std():.2f}")
    
    # Mostrar informaci√≥n temporal si est√° disponible
    if df is not None and 'datetime' in df.columns:
        print(f"\nüìÖ INFORMACI√ìN TEMPORAL:")
        print(f"   ‚Ä¢ Fecha inicial: {df['datetime'].min()}")
        print(f"   ‚Ä¢ Fecha final: {df['datetime'].max()}")
        print(f"   ‚Ä¢ Per√≠odo total: {(df['datetime'].max() - df['datetime'].min()).days} d√≠as")
    
    # Crear histograma b√°sico
    plt.figure(figsize=(14, 5))
    
    plt.subplot(1, 2, 1)
    plt.hist(amplitudes, bins=50, alpha=0.7, color='skyblue', edgecolor='black')
    plt.xlabel('Amplitud')
    plt.ylabel('Frecuencia')
    plt.title('Distribuci√≥n de Amplitudes')
    plt.grid(True, alpha=0.3)
    
    plt.subplot(1, 2, 2)
    plt.hist(amplitudes, bins=50, alpha=0.7, color='lightcoral', edgecolor='black')
    plt.xlabel('Amplitud')
    plt.ylabel('Frecuencia')
    plt.title('Distribuci√≥n de Amplitudes (Escala Log)')
    plt.yscale('log')
    plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
else:
    print("‚ùå Error al cargar los datos")

# %% [markdown]
# ## 4. An√°lisis B√°sico - Todos los Datos

# %%
print("üî¨ AN√ÅLISIS B√ÅSICO - TODOS LOS DATOS")
print("="*50)

# Par√°metros b√°sicos
basic_results = analyzer.analyze_mass_index(
    file_type='csv',
    period='anual',
    min_amplitude=None,  # Sin filtro de amplitud m√≠nima
    use_binning=False,   # Sin binning
    save_results=True,
    save_plots=True
)

# %% [markdown]
# ## 5. An√°lisis Comparativo con Diferentes Par√°metros

# %% [markdown]
# ### 5.1 Efecto del Filtrado por Amplitud M√≠nima

# %%
print("üîç AN√ÅLISIS: EFECTO DEL FILTRADO POR AMPLITUD M√çNIMA")
print("="*55)

# Diferentes valores de amplitud m√≠nima para comparar
min_amplitudes = [None, 1.0, 2.0, 5.0, 10.0]
amplitude_results = {}

for min_amp in min_amplitudes:
    print(f"\n--- Amplitud m√≠nima: {min_amp if min_amp else 'Sin filtro'} ---")
    
    result = analyzer.analyze_mass_index(
        file_type='csv',
        period='anual',
        min_amplitude=min_amp,
        use_binning=False,
        save_results=False,  # No guardar para comparaciones
        save_plots=True
    )
    
    if result and 'todos_los_datos' in result:
        amplitude_results[f"min_{min_amp if min_amp else 'none'}"] = result['todos_los_datos']

# Crear tabla comparativa
if amplitude_results:
    print("\nüìã TABLA COMPARATIVA - EFECTO DE AMPLITUD M√çNIMA:")
    print("-" * 80)
    print(f"{'Filtro':<12} {'N_datos':<8} {'√çndice_s':<10} {'R¬≤':<8} {'P-value':<12}")
    print("-" * 80)
    
    for key, result in amplitude_results.items():
        min_val = key.replace('min_', '').replace('none', 'Sin filtro')
        print(f"{min_val:<12} {result['n_data_points']:<8} "
              f"{result['mass_index']:<10.4f} {result['r_squared']:<8.4f} "
              f"{result['p_value']:<12.2e}")

# %% [markdown]
# ### 5.2 Efecto del Binning Logar√≠tmico

# %%
print("\nüîç AN√ÅLISIS: EFECTO DEL BINNING LOGAR√çTMICO")
print("="*45)

# Comparar con y sin binning
binning_configs = [
    {'use_binning': False, 'n_bins': 0, 'name': 'Sin binning'},
    {'use_binning': True, 'n_bins': 25, 'name': 'Binning 25'},
    {'use_binning': True, 'n_bins': 50, 'name': 'Binning 50'},
    {'use_binning': True, 'n_bins': 100, 'name': 'Binning 100'}
]

binning_results = {}

for config in binning_configs:
    print(f"\n--- {config['name']} ---")
    
    result = analyzer.analyze_mass_index(
        file_type='csv',
        period='anual',
        min_amplitude=None,
        use_binning=config['use_binning'],
        n_bins=config['n_bins'],
        save_results=False,
        save_plots=False
    )
    
    if result and 'todos_los_datos' in result:
        binning_results[config['name']] = result['todos_los_datos']

# Tabla comparativa binning
if binning_results:
    print("\nüìã TABLA COMPARATIVA - EFECTO DEL BINNING:")
    print("-" * 80)
    print(f"{'M√©todo':<15} {'N_regres':<10} {'√çndice_s':<10} {'R¬≤':<8} {'P-value':<12}")
    print("-" * 80)
    
    for name, result in binning_results.items():
        print(f"{name:<15} {result['n_regression_points']:<10} "
              f"{result['mass_index']:<10.4f} {result['r_squared']:<8.4f} "
              f"{result['p_value']:<12.2e}")

# %% [markdown]
# ## 6. An√°lisis Temporal por Per√≠odos

# %% [markdown]
# ### 6.1 An√°lisis Mensual

# %%
print("üìÖ AN√ÅLISIS TEMPORAL - POR MESES")
print("="*35)

monthly_results = analyzer.analyze_mass_index(
    file_type='csv',
    period='mensual',
    min_amplitude=None,
    use_binning=False,
    save_results=True,
    save_plots=True  # No guardar gr√°ficos individuales para evitar saturaci√≥n
)

# Crear gr√°fico resumen mensual
if monthly_results:
    months = []
    mass_indices = []
    r_squared_vals = []
    n_data_points = []
    
    for month, result in monthly_results.items():
        months.append(month)
        mass_indices.append(result['mass_index'])
        r_squared_vals.append(result['r_squared'])
        n_data_points.append(result['n_data_points'])
    
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 10))
    
    # √çndice de masa por mes
    ax1.bar(range(len(months)), mass_indices, color='skyblue', alpha=0.7)
    ax1.set_xlabel('Mes')
    ax1.set_ylabel('√çndice de Masa (s)')
    ax1.set_title('√çndice de Masa por Mes')
    ax1.set_xticks(range(len(months)))
    ax1.set_xticklabels(months, rotation=45)
    ax1.grid(True, alpha=0.3)
    
    # R¬≤ por mes
    ax2.bar(range(len(months)), r_squared_vals, color='lightcoral', alpha=0.7)
    ax2.set_xlabel('Mes')
    ax2.set_ylabel('R¬≤ (Bondad de Ajuste)')
    ax2.set_title('Calidad del Ajuste por Mes')
    ax2.set_xticks(range(len(months)))
    ax2.set_xticklabels(months, rotation=45)
    ax2.grid(True, alpha=0.3)
    
    # N√∫mero de datos por mes
    ax3.bar(range(len(months)), n_data_points, color='lightgreen', alpha=0.7)
    ax3.set_xlabel('Mes')
    ax3.set_ylabel('N√∫mero de Datos')
    ax3.set_title('Cantidad de Datos por Mes')
    ax3.set_xticks(range(len(months)))
    ax3.set_xticklabels(months, rotation=45)
    ax3.grid(True, alpha=0.3)
    
    # Tendencia del √≠ndice de masa
    ax4.plot(range(len(months)), mass_indices, 'o-', color='purple', linewidth=2, markersize=6)
    ax4.set_xlabel('Mes')
    ax4.set_ylabel('√çndice de Masa (s)')
    ax4.set_title('Tendencia Temporal del √çndice de Masa')
    ax4.set_xticks(range(len(months)))
    ax4.set_xticklabels(months, rotation=45)
    ax4.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()

# %% [markdown]
# ### 6.2 An√°lisis Trimestral

# %%
print("üìÖ AN√ÅLISIS TEMPORAL - POR TRIMESTRES")
print("="*38)

quarterly_results = analyzer.analyze_mass_index(
    file_type='csv',
    period='trimestral',
    min_amplitude=None,
    use_binning=False,
    save_results=True,
    save_plots=False
)

# Mostrar tabla resumen trimestral
if quarterly_results:
    print("\nüìã RESUMEN TRIMESTRAL:")
    print("-" * 70)
    print(f"{'Trimestre':<12} {'N_datos':<8} {'√çndice_s':<10} {'R¬≤':<8} {'P-value':<12}")
    print("-" * 70)
    
    for quarter, result in quarterly_results.items():
        print(f"{quarter:<12} {result['n_data_points']:<8} "
              f"{result['mass_index']:<10.4f} {result['r_squared']:<8.4f} "
              f"{result['p_value']:<12.2e}")


"""
# %% [markdown]
# ## 7. An√°lisis Personalizado Interactivo

# %%
print("üõ†Ô∏è SECCI√ìN DE AN√ÅLISIS PERSONALIZADO")
print("="*40)
print("Modifica estos par√°metros seg√∫n tus necesidades:")

# Par√°metros personalizables
CUSTOM_PARAMS = {
    'file_type': 'csv',           # 'csv' o 'txt'
    'period': 'anual',            # 'anual', 'mensual', 'trimestral', etc.
    'min_amplitude': None,        # None o valor num√©rico (ej: 2.0)
    'use_binning': False,         # True o False
    'n_bins': 50,                 # N√∫mero de bins si use_binning=True
    'save_results': True,         # True o False
    'save_plots': True            # True o False
}

print(f"\nPar√°metros actuales:")
for param, value in CUSTOM_PARAMS.items():
    print(f"  ‚Ä¢ {param}: {value}")

print(f"\n¬øQuieres ejecutar el an√°lisis con estos par√°metros? (Ejecuta la siguiente celda)")

# %%
# EJECUTAR AN√ÅLISIS PERSONALIZADO
print("üöÄ EJECUTANDO AN√ÅLISIS PERSONALIZADO...")
print("="*45)

custom_results = analyzer.analyze_mass_index(**CUSTOM_PARAMS)

if custom_results:
    print(f"\n‚úÖ An√°lisis completado exitosamente!")
    print(f"   ‚Ä¢ {len(custom_results)} per√≠odo(s) analizado(s)")
    
    # Mostrar resumen de todos los per√≠odos
    if len(custom_results) > 1:
        print(f"\nüìä RESUMEN DE TODOS LOS PER√çODOS:")
        print("-" * 80)
        print(f"{'Per√≠odo':<15} {'N_datos':<8} {'√çndice_s':<10} {'R¬≤':<8} {'Rango_Amp':<15}")
        print("-" * 80)
        
        for period_name, result in custom_results.items():
            amp_range = f"{result['amplitude_range'][0]:.1f}-{result['amplitude_range'][1]:.1f}"
            print(f"{period_name:<15} {result['n_data_points']:<8} "
                  f"{result['mass_index']:<10.4f} {result['r_squared']:<8.4f} {amp_range:<15}")
else:
    print("‚ùå Error en el an√°lisis personalizado")
"""
"""
# %% [markdown]
# ## 8. Comparaci√≥n de M√©todos (Opcional)

# %%
print("‚öñÔ∏è COMPARACI√ìN DE M√âTODOS")
print("="*30)
print("Comparando diferentes enfoques para el mismo conjunto de datos...")

# Configuraciones para comparar
comparison_configs = [
    {'name': 'Est√°ndar', 'min_amplitude': None, 'use_binning': False},
    {'name': 'Filtrado (min=2)', 'min_amplitude': 2.0, 'use_binning': False},
    {'name': 'Binning 50', 'min_amplitude': None, 'use_binning': True, 'n_bins': 50},
    {'name': 'Filt+Bin', 'min_amplitude': 2.0, 'use_binning': True, 'n_bins': 50}
]

comparison_results = {}

for config in comparison_configs:
    print(f"\n--- M√©todo: {config['name']} ---")
    
    params = {
        'file_type': 'csv',
        'period': 'anual',
        'min_amplitude': config.get('min_amplitude'),
        'use_binning': config.get('use_binning', False),
        'n_bins': config.get('n_bins', 50),
        'save_results': False,
        'save_plots': False
    }
    
    result = analyzer.analyze_mass_index(**params)
    if result and 'todos_los_datos' in result:
        comparison_results[config['name']] = result['todos_los_datos']

# Tabla final de comparaci√≥n
if comparison_results:
    print(f"\nüìã TABLA COMPARATIVA FINAL:")
    print("=" * 90)
    print(f"{'M√©todo':<15} {'N_datos':<8} {'N_regres':<9} {'√çndice_s':<10} {'R¬≤':<8} {'Error_std':<10}")
    print("=" * 90)
    
    for method, result in comparison_results.items():
        print(f"{method:<15} {result['n_data_points']:<8} {result['n_regression_points']:<9} "
              f"{result['mass_index']:<10.4f} {result['r_squared']:<8.4f} {result['std_error']:<10.4f}")
    
    print("=" * 90)
"""
# %% [markdown]
# ## 9. Conclusiones y Pr√≥ximos Pasos

# %%
print("üìù RESUMEN DEL AN√ÅLISIS")
print("="*25)

if 'basic_results' in locals() and basic_results:
    main_result = list(basic_results.values())[0]
    print(f"‚úÖ RESULTADO PRINCIPAL:")
    print(f"   ‚Ä¢ √çndice de masa calculado: s = {main_result['mass_index']:.4f}")
    print(f"   ‚Ä¢ Calidad del ajuste (R¬≤): {main_result['r_squared']:.4f}")
    print(f"   ‚Ä¢ N√∫mero de datos utilizados: {main_result['n_data_points']}")
    print(f"   ‚Ä¢ Rango de amplitudes: {main_result['amplitude_range'][0]:.2f} - {main_result['amplitude_range'][1]:.2f}")

print(f"\nüìÇ ARCHIVOS GENERADOS:")
print(f"   ‚Ä¢ Resultados guardados en: {analyzer.reports_dir}")
print(f"   ‚Ä¢ Gr√°ficos guardados en: {analyzer.figures_dir}")

# %% [markdown]
# ---
# **Fin del an√°lisis**
# 
# Este notebook proporciona un an√°lisis completo del √≠ndice de masa de meteoritos. Los resultados y gr√°ficos se han guardado autom√°ticamente en las carpetas correspondientes del proyecto.
```
